```python
import re

polarity_score = []
texts = list(tweets.text)

for item in texts:
    s = ""
    for word in item.split(" "):
        s = s + ' ' + re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", word)
    tb = TextBlob(s)
    polarity_score.append(tb.polarity)
    
polarity_score
```

```python
polarity_score = pd.Series(polarity_score,name = "Polarity")
tweets = pd.concat([tweets,polarity_score],axis=1)
tweets
```

```python
import numpy as np
tweets["sentiment"] = np.where (tweets.Polarity == 0, "Neutral", np.where(tweets.Polarity < 0, "Negative","Positive"))
tweets
```
