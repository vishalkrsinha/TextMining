```python
## nltk: Natural Language Toolkit used for Natural Language Understanding...

##Other options of nltk package for NLU: TextBlob, gensim, spacey

import nltk.classify.util
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize #Basically used for ngrams
```

```python
# we have a sentence, and we want to extract all the words from it.
sentence = 'Awesome device, and hats off to amazon for such speedy elivery. I ordered it on 16th May 12PM and it got delivered by 17th May 10AM. And so far I m absolutely loving this device, buttery smooth and blazing fast. But its a bit big, afterall 6.7 inches is quite a number and super slippery if used without the case. And I am awestruck by the speed it charges.'

#Text-mining: Step 1: Standardization
sentence = sentence.lower()
sentence
```

```python
#Text-mining: Step 2: Remove special characters
import re

pattern = re.compile(r"[^\w\s]")
sentence = pattern.sub("",sentence)
sentence
```

```python
sentence.split()
```

```python
sentence
```

```python
# We can split the function on a space (" ") to get all the words. 
# However, The problem with this is, we cannot extract punctuation marks like full stops, 
# and this simple parser will not be able to handle every single type of sentence.

# Which is why we should use the word tokenizer provided by the NLTK library. This correctly identifies punctuation marks:

#Text-mining: Step 3: Tokenization
token = word_tokenize(sentence)
print(token)
```

```python
len(token)
```

```python
# Nltk comes inbuilt with a list of stop words for all main languages. 
# To see the stop words for English:

sw = stopwords.words("english")
sw
```

```python
#Text-mining: Step 4: Remove StopWords

clean_tokens = []
for i in token:
    if i not in sw:
        clean_tokens.append(i)
        
clean_tokens
```

```python
len(clean_tokens)
```

# POS Tagging

Remove all punct Identify the parts of speech

```python
# 
# Now, let's get a tag associated with each and every token and see what part of speech these are.
# Whether they're noun, pronoun, adverb, adjective etc.

# By doing so, we can learn more about the constituents of a statement/tweet and see what kind of words present in it.

nltk.pos_tag(clean_tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

#Learn the vocaabulary of the training data
vect= CountVectorizer()
vect
```

```python
vect.fit(clean_tokens)
```

```python
vect.get_feature_names()
```

```python
len(vect.get_feature_names())
```

```python
#Transform training data into a DTM (document term matrix)
simpleTrainDTM = vect.transform(clean_tokens)
simpleTrainDTM
```

```python
simpleTrainDTM.todense()
```

```python
#What does that mean??
import pandas as pd
cv = pd.DataFrame(simpleTrainDTM.todense())
cv.columns = vect.get_feature_names()
cv
```
